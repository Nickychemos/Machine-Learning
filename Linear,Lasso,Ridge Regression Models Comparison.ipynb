### Encoding categorical columns

```python
#Importing neccesary libraries
from sklearn.preprocessing import LabelEncoder
```

```python
#instatiating the label encoder ---> Same as creating a new variable container that will hold the label encoder

labelencoder = LabelEncoder()
```

```python
# fitting and transforming the view column and seeing the categories that were created

df['view_encoded'] = labelencoder.fit_transform(df['view'])

x = df['view'].unique()

categorical_mapping = {category: i for i, category in enumerate(x)}
print(categorical_mapping)
```

```python
# fitting and transforming the condition column and seeing the categories that were created

df['condition_encoded'] = labelencoder.fit_transform(df['condition'])

y = df['condition'].unique()
categorical_mapping = {category: i for i, category in enumerate(y)}
categorical_mapping
```

```python
# fitting and transforming the waterfront column and seeing the categories that were created

df['waterfront_encoded'] = labelencoder.fit_transform(df['waterfront'])

z = df['waterfront'].unique()

categorical_mapping = {category: i for i, category in enumerate(z)}
categorical_mapping
```

### Linear Regression

```python
# Importing the necessary libraries

from sklearn.model_selection import train_test_split #trains, tests and splits the model
from sklearn.linear_model import LinearRegression, Ridge, Lasso # for linear regression
from sklearn.metrics import r2_score # for checking the accuracy of our model
```

```python
df.info()
```

#### Feature Engineering

```python
#For selecting the dependent and independent variables for our models

#defining our x (independent) and y (dependent) variables for our model
x = df[['condition_encoded', 'waterfront_encoded', 'view_encoded','sqft_lot15','sqft_living15','long','lat','zipcode','yr_renovated','yr_built','sqft_basement','sqft_above','grade','floors','sqft_lot','sqft_living','bathrooms','bedrooms']]
y = df['price']

#splitting our dataset into train and test 

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)

#Creating a variable name for our model

model = LinearRegression()

#Train the model of the data

model.fit(x_train, y_train)

#Make predictions

y_pred = model.predict(x_test)

#Calculate the accuracy
r2 = r2_score(y_test, y_pred)

#printing the accuracy score

print('Accuracy: ', r2*100, '%')
```

```python
#Using the Ridge Model

from sklearn.linear_model import Ridge

#defining our x (independent) and y (dependent) variables for our model
x = df[['condition_encoded', 'waterfront_encoded', 'view_encoded','sqft_lot15','sqft_living15','long','lat','zipcode','yr_renovated','yr_built','sqft_basement','sqft_above','grade','floors','sqft_lot','sqft_living','bathrooms','bedrooms']]
y = df['price']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)

#Model Allocation
alpha  = 1.0
model = Ridge(alpha = alpha)

#Training our model
model.fit(x_train, y_train)

#Making predictions
y_pred = model.predict(x_test)

#Getting our accuracy score
r2 = r2_score(y_pred, y_test)
print(f'Accuracy: {r2*100} %')
```

```python
#Using Lasso model

from sklearn.linear_model import Lasso

#defining our x (independent) and y (dependent) variables for our model
x = df[['condition_encoded', 'waterfront_encoded', 'view_encoded','sqft_lot15','sqft_living15','long','lat','zipcode','yr_renovated','yr_built','sqft_basement','sqft_above','grade','floors','sqft_lot','sqft_living','bathrooms','bedrooms']]
y = df['price']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)

#Model Allocation
alpha = 1.0
model = Lasso(max_iter=5000, alpha = alpha)

#Training the model
model.fit(x_train,y_train)

#Making predictions
y_prediction = model.predict(x_test)

#Checking the accuracy of our score
r2 = r2_score(y_prediction, y_test)
print(f' Accuracy: {r2*100} %')
```

```python
#It can be concluded that the linear regression model performed better than the Lasso and Ridge regression models.
```
