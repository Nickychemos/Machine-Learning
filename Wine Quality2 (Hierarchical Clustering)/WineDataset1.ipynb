```python
#Loading the wine dataset

from sklearn.datasets import load_wine

data = load_wine()
data
```

```python
#Choosing the X variables (Labelled data above)

X = data.data
X
```

# Hierarchical Clustering 

```python
#Importing the necessary libraries

import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage,dendrogram,fcluster
from sklearn.metrics import silhouette_score
```

```python
#Creating a linkage matrix

linkage_matrix = linkage(X,'ward')
print(linkage_matrix[:5])
```

```python
#Visualizing the matrix using a dendrogram

plt.figure(figsize=(20,10))
dendrogram(linkage_matrix)
plt.xlabel('Clusters')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
#Creating the clusters

clusters = fcluster(linkage_matrix,2,criterion='maxclust')
print(clusters)
```

```python
#Scatter plot to see the clusters

plt.figure(figsize=(8,6))
plt.scatter(X[:,1], X[:,6], c=clusters)
plt.xlabel('malic_acid')
plt.ylabel('flavanoids')
plt.show()
```

```python
#Getting the silhoette score of the model

sil_score = silhouette_score(X, clusters)
print('The silhouette score of the Hierarchical Cluster is ', round(sil_score,2))
```
